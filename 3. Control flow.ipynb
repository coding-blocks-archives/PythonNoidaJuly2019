# 2. Looping statements

## (i) while

The while statement allows you to repeatedly execute a block of statements as long as a condition is true.

![](images/while-loop.png)

````python
a = 10
while a > 0:
    print(a)
    a-=2
````

````python
while True:
    print("I'll not stop")
````

## (ii) for

The for..in statement is another looping statement which iterates over a sequence of objects i.e. go through each item in a sequence.

![](images/for_loop.jpg)

````python
name = "python"
````

```py
for iterator in iterable:
    body
```

````python
for char in name:
    print(char)
````

````python
# for (int i=1; i<10; i++) {
#     cout << i;
#     i = 100;
# }
for i in range(1, 10, 2):
    print(i)
    i = 100
````

## Nested loops

Yes! Loops can also be nested.

Lets do matrix addition!

````python
n = 5
for i in range(n):
    for j in range(n):
        print("* ", end = "")
    print()
````

````python
print("Hello")
print("World")
````

````python
n = 5
for i in range(n):
    for j in range(n-i-1):
        print("  ", end = "")
    for j in range(i+1):
        print("* ", end = "")
    print()
````
