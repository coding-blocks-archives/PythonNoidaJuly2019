```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.amazon.in/s"
params = {
    "k": "beaan bag"
}
r = requests.get(url, params = params)
```

```python
r.url
```

```python
soup = BeautifulSoup(r.content)
```

```python
products = soup.findAll('div', attrs = {"class": "s-result-item"})
```

```python
products[0].find('div', attrs={"class": "product-price-row"})
```

```python
from IPython.display import Image, display
```

```python
for product in products:
    title = product.find('p', attrs = {"class": "product-title"}).text
    print(title)
    price = product.find('span', attrs={"class": "product-desc-price"}).text
    print(price)
    actual_price = product.find('span', attrs={"class": "product-price"}).text
    print(actual_price)
    imgtag = product.find('img')
    if "src" in imgtag.attrs:
        display(Image(imgtag.attrs["src"]))
    elif "data-src" in imgtag.attrs:
        display(Image(imgtag.attrs["data-src"]))
```

```python
discounts = []
```

```python
for i in range(len(products)):
    discounts.append(
        products[i].find('div', attrs = {"class": "product-discount"}).text
    )
```
